```python
person = ["Hira Ahmed",35,3.6,["java","python","c++",["dell","HP"]]]
person[3][3][1]
```

```python
person[0][3][3]
```

```python
# Value at index 0 from the first outer list
value_from_firstout = person[0]
print(f"Value from first outer list: {value_from_firstout}")

# Value at index 0 from the first inner list
value_from_first = person[3][0]
print(f"Value from first inner list: {value_from_first}")

# Value at index 3 from the second inner list
value_from_second = person[3][3][1]
print(f"Value from second inner list: {value_from_second}")
```
