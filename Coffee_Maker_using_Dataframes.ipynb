```python
import numpy as np
import pandas as pd

# Define the dictionary
coffee_maker = {
    'No.': list(range(1, 7)),
    'Menu': ["Café au Lait", "Café Bombón", "Café Cubano", "Café con Leche", "Café del Tiempo", "Café de Olla"],
    'Category': np.random.choice(['Espresso', 'Americano', 'Latte', 'Cappuccino', 'Mocha', 'Macchiato', 'Flat White', 'Iced Coffee'], 6),
    'Price': np.random.choice([5.99, 8.99, 10.00, 15.99], 6),
    'Description': np.random.choice(['Hot', 'Cold'], 6),
    'Size': np.random.choice(['Small', 'Medium', 'Large'], 6),
    'Extras': np.random.choice(['Extra Shot', 'Flavor Syrup', 'Almond Milk', 'Oat Milk', 'Soy Milk'], 6),
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(coffee_maker)

# Greeting Function
def greetings():
    line1 = "===================================="
    welcome_message = "WELCOME TO COFFEE MAKER"
    line3 = "===================================="
    return f"{line1}\n{welcome_message}\n{line3}"

# Menu Function
def coffee_menu():
    quest1 = "Please select your coffee:"
    return f"{quest1}\n{df}"

# Coffee Selection Function
def select_coffee():
    selection = input("\nEnter number of your choice: ")
    if not selection.isdigit():
        print("Invalid input. Please enter a number.")
        return select_coffee()

    selection = int(selection)
    if selection not in df['No.'].values:
        print("Invalid selection. Please try again.")
        return select_coffee()

    selected_coffee = df[df['No.'] == selection].to_dict('records')[0]
    return selected_coffee

# User Order Function
def user_order(selected_coffee):
    print("\n=============YOUR ORDER=============")
    print(f"You have selected {selected_coffee['Menu']} ({selected_coffee['Category']}).")
    print(f"Description: {selected_coffee['Description']}")
    print(f"Size: {selected_coffee['Size']}")
    print(f"Extras: {selected_coffee['Extras']}")
    print(f"Your Total Bill is: PKR. {selected_coffee['Price']:.2f}")

# Order Complete or Closing Function
def closing():
    line1 = "===================================="
    signoff = "ENJOY YOUR COFFEE!"
    return f"{line1}\n{signoff}"

# Main Program
def process():
    print(greetings())
    print(coffee_menu())
    coffee_choice = select_coffee()  # Select coffee
    user_order(coffee_choice)
    print(closing())

# Start the program
process()
```
